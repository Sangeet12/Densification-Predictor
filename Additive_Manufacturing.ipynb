```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/3D Printing Project - Sheet1.csv')
```

```python
print(data)
```

```python
data = data.sample(frac = 1)
print(data)
```

```python
print(data['Material'].unique())
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
data['Material']= label_encoder.fit_transform(data['Material'])
print(data)
```

```python
x = data.drop(['Densification'], axis = 1)
print(x)
```

```python
y = data[['Densification']]
print(y)
```

```python
x = x.astype(float)
y = y.astype(float)
print(x.dtypes)
print(y.dtypes)
```

```python
x = x.to_numpy()
y = y.to_numpy()
```

```python
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x = train_test_split(x, test_size=0.2)
train_y, test_y = train_test_split(y, test_size=0.2)
```

```python
from keras import models
from keras import layers
from keras import regularizers
from keras.layers import Dense, Activation, Flatten
```

```python
model = models.Sequential()
model.add(Dense(128, kernel_initializer='normal',input_dim = train_x.shape[1], activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mape'])
```

```python
history = model.fit(train_x, train_y, epochs = 500, batch_size = 50)
```

```python
pred_test = model.predict(test_x)
pred_train = model.predict(train_x)
```

```python
from keras import losses
```

```python
mape = losses.MeanAbsolutePercentageError()
```

```python
test_error = mape(test_y,pred_test)
train_error = mape(train_y, pred_train)
```

```python
print("Training Error: ", train_error)
print("Testing Error: ", test_error)
```
